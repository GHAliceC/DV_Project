```python
import pandas as pd
#import matplotlib.pyplot as plt
#import IPython.display
import ipywidgets as pget
from bqplot import pyplot as bqplt
import bqplot as bq
import numpy as np
```

```python
magnitude = pd.read_table("data-readonly/transportable_array/data_tohoku_norm_transpose.csv",header = None,delimiter=",")
location = pd.read_table("data-readonly/transportable_array/location.txt", header = None)
```

```python
location_df = pd.concat([pd.DataFrame([range(len(location[0]))]).transpose(),location], axis=1)
location_df.columns = ["idx",'x', 'y', '3', '4']
```

```python
# Funtion to check the whether the detecter is selected or not
def checkSelect(selected):
    if(len(selected) == 0): 
        #print(0)
        return(0) # if not selected return 0
    else:   
        #print(selected[0])
        return(selected[0]) # else return the selected result
```

```python
# figure = bqplt.figure("site")
site_location = bqplt.scatter(x=location_df['x'], y =location_df['y'],
                              interactions = {'click':'select'},unselected_style = {'opacity':0.4}, name = location_df['idx'],
                            )
site_locationFig = bq.pyplot.figure(key="site_location",marks=[site_location], axes = [axsc_x, axsc_y],
             title='Detactor location')


waveplot = bq.pyplot.plot(x = np.arange(len(magnitude.loc[0])) , 
                           y=magnitude.loc[checkSelect(site_location.selected)] )
waveplotFig =bq.pyplot.figure(key= "waveplot",marks=[waveplot], title='Wave plot')

site_locationFig

```
