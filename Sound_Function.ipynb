```python
import matplotlib.pyplot as plt
import matplotlib as mp
import numpy as np
import IPython
import pandas as pd
```

```python
# the magnitutde original data
data = pd.read_csv('data-readonly/transportable_array/data_tohoku_norm_transpose.csv', header = None)
```

```python

# lenth: how long we want to use the original data
# freq: frequency of sleep
# rate: frequncy rate of the rate keyword argument in the 
# data: the magnitude dataframe 

def genSoundWave(data,lenth = 4000, freq = 0.02 , siteNum = 0, rate = 7000 ):
    part = lenth*freq # the number of interval we partition the data == length(sec) of the video
    soundTest = np.split(data[siteNum][0:lenth], part) # split the data 2D array of each element gonna be repeat
    partLen = lenth/part # number of data within each partition
    repeatTime = rate/partLen # how many times do we need to repeat until it reach the rate
    soundOut = np.repeat(np.array(soundTest), repeatTime , axis= 0) # repeat each interval repeatTime
    finalWave = soundOut.flatten() # flatten the 2D array into 1D
    return(finalWave)
```

```python
SelectedSite = 400
finalWave = genSoundWave(data, siteNum= SelectedSite)
IPython.display.Audio(data = finalWave, rate = 7000)
```
